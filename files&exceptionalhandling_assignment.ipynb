**Ques 1**- What is the difference between interpreted and compiled languages?

**Ques 9**- What is the difference between multithreading and multiprocessing?

**Ans 9**- **Multithreading** runs multiple threads (lightweight, shared memory) within a single process, ideal for I/O-bound tasks like network requests. **Multiprocessing** uses separate memory spaces and multiple CPU cores for true parallelism, best for CPU-bound tasks like heavy calculations. Multithreading is faster to create but limited by the Global Interpreter Lock (GIL) in Python, whereas multiprocessing offers better fault isolation.

```python
# Ques 4-  Write a Python script that reads from one file and writes its content to another file?
# Answer
def copy_file_simple(source_filename, destination_filename):
  try:
    with open(source_filename, 'r') as f_source:
      content = f_source.read()
    with open(destination_filename, 'w') as f_dest:
      f_dest.write(content)
      print(f"Successfully copied contents from '{source_filename}' to '{destination_filename}'")
  except FileNotFoundError:
      print(f"Error: One of the files was not found.")
```

```python
# Ques 5- How would you catch and handle division by zero error in Python?
# Answer
def safe_division(numerator, denominator):
    try:
      result = numerator / denominator
      return result
    except ZeroDivisionError:
        print("Error: Cannot divide by zero.")
        return None
    except TypeError:
        print("Error: Both inputs must be numbers.")
        return None
    except Exception as e:
        print(f"An unexpected error occurred: {e}")
        return None

# --- Usage Examples ---

# Case 1: Successful division
print(f"10 / 2 = {safe_division(10, 2)}")

# Case 2: Division by zero (handled gracefully)
print(f"10 / 0 = {safe_division(10, 0)}")

# Case 3: A different error (handled by TypeError or generic Exception)
print(f"10 / 'a' = {safe_division(10, 'a')}")
```

```python
# Ques 8-Write a program to handle a file opening error using exception handling?
# Answer
def open_file_safely(filename):
  try:
    with open(filename, 'r') as file:
      print(f"Successfully opened '{filename}'. Reading contents:")
      contents = file.read()
      print(contents)
  except FileNotFoundError:
    print(f"Error: The file '{filename}' was not found.")
  except IOError as e:
    print(f"An I/O error occurred: {e}")
  except Exception as e:
    print(f"An unexpected error occurred: {e}")
  else:
    print(f"Finished processing '{filename}' without exceptions.")
  finally:
    print("Closing file operation.")
```

```python
# Ques 12-  Write a program that demonstrates using multiple except blocks to handle different types of exceptions?
# Answer
def safe_division():
  try:
    first_number_input = input("Enter the numerator: ")
    second_number_input = input("Enter the denominator: ")
    first = float(first_number_input)
    second = float(second_number_input)
    result = first / second
    print(f"{first} divided by {second} is {result}")

  except ValueError:
    print("Error: You must enter a valid number.")
  except ZeroDivisionError:
    print("Error: You can't divide by zero.")

  except Exception as e:
    print(f"An unexpected error occurred: {e}")

  else:
    print("Division performed successfully without any errors.")

  finally:
    print("Execution of the try-except block is complete.")

safe_division()
```

```python
# Ques 14- Write a program that uses the logging module to log both informational and error messages?
# Answer
import logging
import sys
logging.basicConfig(
    level=logging.INFO,
    format="%(asctime)s - %(levelname)s - %(name)s - %(message)s",
    handlers=[logging.StreamHandler(sys.stdout),logging.FileHandler("app.log")])
logger = logging.getLogger(__name__)
def perform_operation(value):
  logger.info(f"Attempting operation with value: {value}")
  try:
    result = 10 / value
    logger.info(f"Operation successful, result: {result}")
    return result
  except ZeroDivisionError as e:
    logger.error(f"Error: Division by zero is invalid for value: {value}", exc_info=True)
    return None
  except Exception as e:
    logger.exception(f"An unexpected exception occurred with value: {value}")
  return None

if __name__ == "__main__":
    logger.info("Program started.")

    perform_operation(5)

    perform_operation(0)

    perform_operation(2)

    logger.info("Program finished.")
```

```python
# Ques 15- Write a Python program that prints the content of a file and handles the case when the file is empty?
# Answer
import os

def print_file_content(file_path):
  try:
    with open(file_path, 'r') as file:
      content = file.read()
      if not content.strip():
        print(f"'{file_path}' is an empty file or contains only whitespace.")
      else:
        print(f"Content of '{file_path}':")
        print(content)
  except FileNotFoundError:
    print(f"Error: The file '{file_path}' was not found.")
  except Exception as e:
    print(f"An unexpected error occurred: {e}")

non_empty_file = "non_empty.txt"
with open(non_empty_file, "w") as f:
    f.write("Hello, world!\nThis file has content.")

print_file_content(non_empty_file)
print("-" * 20)

empty_file = "empty.txt"
with open(empty_file, "w") as f:
    pass

print_file_content(empty_file)
print("-" * 20)

whitespace_file = "whitespace.txt"
with open(whitespace_file, "w") as f:
    f.write("   \n\n")

print_file_content(whitespace_file)
print("-" * 20)

print_file_content("non_existent.txt")
```

```python
# Ques 17-Write a Python program to create and write a list of numbers to a file, one number per line?
# Answer
def write_numbers_to_file(filename, numbers):
  try:
    with open(filename, 'w') as file:
      for number in numbers:
        file.write(str(number) + '\n')
        print(f"Successfully wrote numbers to {filename}")
  except IOError as e:
    print(f"Error writing to file {filename}: {e}")

if __name__ == "__main__":
    data_list = [10, 20, 30, 45, 60, 75, 90]
    file_name = "output_numbers.txt"

    write_numbers_to_file(file_name, data_list)
    print(f"\n--- Contents of {file_name} ---")
    try:
        with open(file_name, 'r') as file:
            print(file.read())
    except IOError as e:
        print(f"Error reading file {file_name}: {e}")
```

```python
# Ques 21-Write a Python program that reads a file and prints the number of occurrences of a specific word?
# Answer
import re

def count_word_occurrences(filename, word_to_find):
  count = 0
  word_to_find = word_to_find.lower()

  try:
    with open(filename, 'r', encoding='utf-8') as file:
      content = file.read().lower()
      matches = re.findall(r'\b' + re.escape(word_to_find) + r'\b', content)
      count = len(matches)

  except FileNotFoundError:
        print(f"Error: The file '{filename}' was not found.")
        return 0
  except Exception as e:
        print(f"An error occurred: {e}")
        return 0
  return count

if __name__ == "__main__":
    sample_file_content = """
    Hello world, this is a sample text file.
    It has the word sample multiple times, sample.
    We are counting sample occurrences in this sample file.
    """
    with open("sample.txt", "w", encoding='utf-8') as f:
        f.write(sample_file_content.strip())
        file_name = "sample.txt"
        word_to_count = "sample"

    occurrences = count_word_occurrences(file_name, word_to_count)
    print(f"The word '{word_to_count}' appears {occurrences} times in the file '{file_name}'.")

    word_not_in_file = "python"
    occurrences_not_found = count_word_occurrences(file_name, word_not_in_file)
    print(f"The word '{word_not_in_file}' appears {occurrences_not_found} times in the file '{file_name}'.")
```

```python
# Ques 23- Write a Python program that writes to a log file when an error occurs during file handling.
# Answer
import logging
logging.basicConfig(filename='file_errors.log',level=logging.ERROR,format='%(asctime)s - %(levelname)s - %(message)s')

def safe_file_operation(filename, mode='r', content=None):
  try:
    with open(filename, mode) as f:
      if mode == 'r':
        return f.read()
      elif mode in ('w', 'a'):
        f.write(content)
  except FileNotFoundError as e:
    logging.error(f"File not found: {filename}. Details: {e}")
  except IOError as e:
    logging.error(f"I/O error with {filename}. Details: {e}")
  except Exception as e:
    logging.exception(f"Unexpected error: {e}")
  finally:
    pass

safe_file_operation("non_existent_file.txt", mode='r')
```
